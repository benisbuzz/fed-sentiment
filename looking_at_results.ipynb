```python
from api_calls import get_multiple_api_calls
import pandas as pd
import regex as re
import json
```

```python
speeches = list(pd.read_parquet("data/data_scraping/scraped_data/speeches.parquet")["title"])
```

```python
with open("data/prompts/title_filter_prompt.txt", "r") as file:
    classify_prompt = file.read().splitlines()
```

```python
classifications = await get_multiple_api_calls("gpt-4-1106-preview", classify_prompt, speeches)
```

```python
def get_label(string:str):
    match = re.search(r"\b(irrelevant|relevant)\b", string, re.IGNORECASE)
    if match and match.group(0).lower() == "relevant":
        return 1
    return 0
    
```

```python
classifications_filtered = [get_label(ele[0]) for ele in classifications]
```

```python
result = {title:label for title, label in zip(speeches, classifications_filtered)}
```

```python
with open("title_classifications.json", "w") as file:
    json.dump(result, file)
```

```python
with open("title_classifications.txt", "w") as file:
    for title, label in result.items():
        file.write(f"{title} : {label}\n")
```
