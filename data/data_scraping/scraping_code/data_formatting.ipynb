```python
import pandas as pd
import os
import regex as re
from nltk.tokenize import sent_tokenize
```

```python
speech_data = pd.read_parquet("../scraped_data/speeches.parquet")

all_text = []
for index, text in enumerate(speech_data["text"]):
    t = re.sub(r'(\r\n|\r|\n)', ' ', text)
    t = re.sub('  ', ' ', t)
    t = re.sub(r'(?<=[.,;])(?=[^\s])', r' ', t)
    sent_tokens = sent_tokenize(text.replace("\n", ""))
    sent_tokens = [re.sub(r'\s+', ' ', sent) for sent in sent_tokens]
    all_text.append(sent_tokens)

speech_data["text"] = all_text
speech_data.set_index("date", drop=True, inplace=True)
speech_data.to_parquet("../scraped_data/speeches.parquet")
```

```python
select_files = os.listdir("../press_conferences/csv/select/")

df_dict = {"date":[], "chair": [], "split_text":[]}

for filename in select_files:
    data = pd.read_csv("../press_conferences/csv/select/" + filename)
    df_dict["split_text"].append(list(data["sentence"]))
    df_dict["chair"].append(data.iloc[0,1])
    df_dict["date"].append(pd.Timestamp(re.search(r"\d{8}", filename).group()))
    
pc_df = pd.DataFrame(df_dict)
pc_df.set_index("date", inplace=True)
pc_df.sort_index(inplace=True)
pc_df.to_parquet("../scraped_data/press_conferences.parquet")
```
